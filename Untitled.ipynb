```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('NFA 2019 public_data.csv')
df
```

```python
df.dropna(inplace=True)
df.corr()
```

```python
X=df['crop_land','grazing_land']
y=df['built_up_land']
```

```python
lm=LinearRegression()
model=lm.fit(X,y)
```

```python
built_up_land_predict=model.predict()
plt.figure(figsize=(12,6))
plt.plot(df['crop_land'],df['built_up_land'],'o')
plt.plot(df['crop_land'],built_up_land_predict,'r',linewidth=2)
plt.xlabel('cropland')
plt.ylabel('builtland')
plt.title('builtland vs cropland')
plt.show()
                            
                            
```
